## Importing pandas and csv data

```python
import pandas as pd
pd.set_option('display.max_columns', None)
```

```python
manhattan = pd.read_csv('../Data/2022/2022_manhattan.csv')
bronx = pd.read_csv('../Data/2022/2022_bronx.csv')
brooklyn = pd.read_csv('../Data/2022/2022_brooklyn.csv')
queens = pd.read_csv('../Data/2022/2022_queens.csv')
staten_island = pd.read_csv('../Data/2022/2022_staten_island.csv')
```

## Combining the datasets and cleaning them based on above discoveries

```python
boroughs2022 = pd.concat([manhattan,bronx,brooklyn,queens,staten_island])
boroughs2022.dropna(how='all',inplace=True)
boroughs2022.fillna("",inplace=True)

boroughs2022.drop(['EASE-MENT','APARTMENT NUMBER','BUILDING CLASS AT PRESENT','BUILDING CLASS AT TIME OF SALE','ADDRESS'],axis=1,inplace=True);

boroughs2022 = boroughs2022[boroughs2022['SALE PRICE'].str.contains('-',regex=False) == False]
boroughs2022 = boroughs2022[boroughs2022['LAND SQUARE FEET'].str.contains('-',regex=False) == False]
boroughs2022 = boroughs2022[boroughs2022['GROSS SQUARE FEET'].str.contains('-',regex=False) == False]
boroughs2022['SALE PRICE']=boroughs2022['SALE PRICE'].str.replace('$','', regex=False)
boroughs2022['SALE PRICE']=boroughs2022['SALE PRICE'].str.replace(',','', regex=False)
boroughs2022['LAND SQUARE FEET']=boroughs2022['LAND SQUARE FEET'].str.replace(',','', regex=False)
boroughs2022['GROSS SQUARE FEET']=boroughs2022['GROSS SQUARE FEET'].str.replace(',','', regex=False)
boroughs2022 = boroughs2022[boroughs2022['LAND SQUARE FEET'] != '']
boroughs2022 = boroughs2022[boroughs2022['GROSS SQUARE FEET'] != '']
boroughs2022 = boroughs2022.astype({'SALE PRICE': float})
boroughs2022 = boroughs2022.astype({'LAND SQUARE FEET': float})
boroughs2022 = boroughs2022.astype({'GROSS SQUARE FEET': float})
boroughs2022 = boroughs2022.astype({'YEAR BUILT': str})
boroughs2022 = boroughs2022.astype({'ZIP CODE': str})
boroughs2022 = boroughs2022[boroughs2022['SALE PRICE'] > 1000]
boroughs2022['SALE DATE'] = pd.to_datetime(boroughs2022['SALE DATE'])
boroughs2022.reset_index(drop=True,inplace=True)
```

## Making basic graphs to see if we can pull meaningful information out of the dataset

```python
import matplotlib.pyplot as plt
import matplotlib.dates as dates
```

```python
boroughs2022.groupby('BOROUGH')['SALE PRICE'].mean().plot(kind='bar')
plt.xticks(rotation=0)
plt.xlabel("Borough")
plt.ylabel("Average Sale Price ($)")
plt.title("Average Sale Price by NYC Borough")
plt.show()
```

```python
x_data = boroughs2022['SALE DATE']
y_data = boroughs2022['SALE PRICE']

fig, ax = plt.subplots()
ax.scatter(x_data, y_data, alpha = 0.5)

# years1=dates.YearLocator(2,month=1,day=1)     This is useful for a dataset spanning multiple years
plt.xticks(rotation="45")
ax.xaxis.set_major_locator(dates.MonthLocator())
ax.xaxis.set_major_formatter(dates.DateFormatter('%b'))

plt.xlabel("Month")
plt.ylabel("Sale Price ($)")
plt.title("NYC Sale Prices in 2022")

plt.show()
```
