```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
manhattan = pd.read_csv('../Data/2022/2022_manhattan.csv')
bronx = pd.read_csv('../Data/2022/2022_bronx.csv')
brooklyn = pd.read_csv('../Data/2022/2022_brooklyn.csv')
queens = pd.read_csv('../Data/2022/2022_queens.csv')
statenisland = pd.read_csv('../Data/2022/2022_staten_island.csv')
pd.set_option('display.max_columns', None)
```

```python
x = pd.concat([manhattan,bronx,brooklyn,queens,statenisland])
# x.head()
```

```python
x.drop(['EASE-MENT','APARTMENT NUMBER','BUILDING CLASS AT PRESENT','BUILDING CLASS\nAT TIME OF SALE','ADDRESS'],axis=1,inplace=True);
# x.info()
```

```python
x.head()
```

```python
x.columns = x.columns.str.replace(' ', '_')
x.columns = x.columns.str.replace('\n', '')
x.rename(columns={"RESIDENTIALUNITS": "RESIDENTIAL UNITS", "COMMERCIALUNITS": "COMMERCIAL UNITS", "BUILDING CLASSAT TIME OF SALE": "BUILDING CLASS AT TIME OF SALE"},inplace=True)
x['LAND_SQUARE_FEET']=x['LAND_SQUARE_FEET'].str.replace(',','')
x['GROSS_SQUARE_FEET']=x['GROSS_SQUARE_FEET'].str.replace(',','')
x['SALE_PRICE']=x['SALE_PRICE'].str.replace(',','')
x = x.astype({'LAND_SQUARE_FEET': float})
x = x.astype({'GROSS_SQUARE_FEET': float})
x = x.astype({'SALE_PRICE': float})
```
